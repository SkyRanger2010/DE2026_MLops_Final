```python
import os
import requests
from requests.auth import HTTPBasicAuth
from pprint import pprint

# Внутри Jupyter-контейнера используем имя docker-сервиса lakefs.
# Учётные данные берутся из .env (LAKEFS_ACCESS_KEY, LAKEFS_SECRET_KEY).
LAKEFS_ENDPOINT = os.environ.get("LAKEFS_ENDPOINT", "http://lakefs:8000/api/v1")
LAKEFS_ACCESS_KEY = os.environ.get("LAKEFS_ACCESS_KEY", "")
LAKEFS_SECRET_KEY = os.environ.get("LAKEFS_SECRET_KEY", "")

auth = HTTPBasicAuth(LAKEFS_ACCESS_KEY, LAKEFS_SECRET_KEY)
print("LakeFS endpoint:", LAKEFS_ENDPOINT)
```

```python
# Проверка health (через имя сервиса в docker-сети)
health = requests.get("http://lakefs:8000/health", timeout=10)
print("Health status:", health.status_code)
print("Health body:", health.text)
```

```python
# Пример SDK-клиента lakefs-client
from lakefs_client import Configuration, ApiClient
from lakefs_client.api.repositories_api import RepositoriesApi

cfg = Configuration(
    host=LAKEFS_ENDPOINT,
    username=LAKEFS_ACCESS_KEY,
    password=LAKEFS_SECRET_KEY,
)

with ApiClient(cfg) as client:
    repos_api = RepositoriesApi(client)
    repos = repos_api.list_repositories()

print(repos)
```
